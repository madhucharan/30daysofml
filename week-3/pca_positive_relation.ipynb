```python
import numpy as np
```

```python
import scipy as sp
import matplotlib.pyplot as plt
```

```python
mean = np.array([3, 5])
cov = np.array([[1, 0.5], [0.5, 1]])
joint_data = np.random.multivariate_normal(mean=mean, cov=cov, size=2000)
```

```python
plt.scatter(joint_data[:, 0], joint_data[:, 1])
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(2)
```

```python
fitted_data = pca.fit(joint_data)
```

```python
fitted_data.explained_variance_ratio_
```

```python
X = joint_data
pca = fitted_data
plt.scatter(X[:, 0], X[:, 1], alpha=.3, label='samples')
for i, (comp, var) in enumerate(zip(pca.components_, pca.explained_variance_)):
    comp = comp * var  # scale component by its variance explanation power
    plt.plot([3, comp[0] + 3], [5, comp[1] + 5], label=f"Component {i}", linewidth=5,
             color=f"C{i + 2}")
```
